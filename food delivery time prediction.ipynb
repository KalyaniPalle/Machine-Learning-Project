```python
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].apply(lambda x: 5.0 if x>5.0 else x)
```

```python
from geopy.distance import geodesic

for i in range (0,2060):
    restaurant = ((df.at[i,"Restaurant_latitude"]),(df.at[i,"Restaurant_longitude"]))
    delivery = ((df.at[i,"Delivery_location_latitude"]),(df.at[i,"Delivery_location_longitude"]))
    distance = geodesic(restaurant,delivery).m
    df.at[0+i,"distance"] = distance
```

```python
df['speed'] = df['distance'] / df['Time_taken(min)']
```
